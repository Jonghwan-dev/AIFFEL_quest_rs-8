```python
# Import required libraries
import os
import urllib
import cv2
import numpy as np
from pixellib.semantic import semantic_segmentation
from matplotlib import pyplot as plt
```

```python
# Set reference image path
img_path = os.getenv('HOME')+'/aiffel/human_segmentation/images/face.jpg'  
img_orig = cv2.imread(img_path) 

# Print shape and image
print(img_orig.shape)
plt.imshow(cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Get pre-trained model
model_dir = os.getenv('HOME')+'/aiffel/human_segmentation/models' 
model_file = os.path.join(model_dir, 'deeplabv3_xception_tf_dim_ordering_tf_kernels.h5') 

model_url = 'https://github.com/ayoolaolafenwa/PixelLib/releases/download/1.1/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5' 
urllib.request.urlretrieve(model_url, model_file)

model = semantic_segmentation()
model.load_pascalvoc_model(model_file)

segvalues, output = model.segmentAsPascalvoc(img_path)
```

```python
# Declare label names and predict on image
LABEL_NAMES = [
    'background', 'aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus',
    'car', 'cat', 'chair', 'cow', 'diningtable', 'dog', 'horse', 'motorbike',
    'person', 'pottedplant', 'sheep', 'sofa', 'train', 'tv'
]
plt.imshow(output)
plt.show()
segvalues
```

```python
for class_id in segvalues['class_ids']:
    print(LABEL_NAMES[class_id])

colormap = np.zeros((256, 3), dtype = int)
ind = np.arange(256, dtype=int)

for shift in reversed(range(8)):
    for channel in range(3):
        colormap[:, channel] |= ((ind >> channel) & 1) << shift
    ind >>= 3

colormap[:20]
```

```python
# Grayscaled segmentation map
seg_color = (128,128,192)
seg_map = np.all(output==seg_color, axis=-1) 
print(seg_map.shape) 
plt.imshow(seg_map, cmap='gray')
plt.show()
```

```python
# Create mask
img_show = img_orig.copy()

img_mask = seg_map.astype(np.uint8) * 255

color_mask = cv2.applyColorMap(img_mask, cv2.COLORMAP_JET)

img_show = cv2.addWeighted(img_show, 0.6, color_mask, 0.4, 0.0)

plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Create blur
img_orig_blur = cv2.GaussianBlur(img_orig, (63,63), 0)
plt.imshow(cv2.cvtColor(img_orig_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Create blured backgraound
img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)
img_bg_mask = cv2.bitwise_not(img_mask_color)
img_bg_blur = cv2.bitwise_and(img_orig_blur, img_bg_mask)
plt.imshow(cv2.cvtColor(img_bg_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Concatenate
img_concat = np.where(img_mask_color==255, img_orig, img_bg_blur)
plt.imshow(cv2.cvtColor(img_concat, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Import required libraries
import os
import urllib
import cv2
import numpy as np
from pixellib.semantic import semantic_segmentation
from matplotlib import pyplot as plt
```

```python
def bgblur(img_path=os.getenv('HOME')+'/aiffel/human_segmentation/images/face.jpg'):

    # Create grid for plots
    f, axarr = plt.subplots(2,2)

    # Set and print reference image 
    img_orig = cv2.imread(img_path)
    axarr[0, 0].imshow(cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB))
    axarr[0, 0].set_title('Reference Image')

    # Get pre-trained model
    model_dir = os.getenv('HOME') + '/aiffel/human_segmentation/models' 
    model_file = os.path.join(model_dir, 'deeplabv3_xception_tf_dim_ordering_tf_kernels.h5') 
    model = semantic_segmentation()
    model.load_pascalvoc_model(model_file)
    segvalues, output = model.segmentAsPascalvoc(img_path)

    # Declare label names and predict on image
    LABEL_NAMES = [
        'background', 'aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus',
        'car', 'cat', 'chair', 'cow', 'diningtable', 'dog', 'horse', 'motorbike',
        'person', 'pottedplant', 'sheep', 'sofa', 'train', 'tv'
    ]
    axarr[0, 1].imshow(output)
    axarr[0, 1].set_title('Predicted Segments')
    # 디버깅 과정 기록
    # Predict 값을 출력하기 위해 output이라는 변수이름을 사용했는데 나중에
    # np.all(output==seg_color, axis=-1)에서 output을 변수이름으로 읽는 일이 일어나 시간이 많이 소요되었다.
    out = ', '.join(f'"{LABEL_NAMES[class_id]}"' for class_id in segvalues['class_ids'])
    print(f'This image contains {out}.')

    # Create mask
    seg_color = (128,128,192)
    seg_map = np.all(output==seg_color, axis=-1) 
    img_show = img_orig.copy()
    img_mask = seg_map.astype(np.uint8) * 255
    color_mask = cv2.applyColorMap(img_mask, cv2.COLORMAP_JET)
    img_show = cv2.addWeighted(img_show, 0.6, color_mask, 0.4, 0.0)
    axarr[1, 0].imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
    axarr[1, 0].set_title('Masked Image')

    # Blur backgraound
    img_orig_blur = cv2.GaussianBlur(img_orig, (63,63), 0)
    img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)
    img_bg_mask = cv2.bitwise_not(img_mask_color)
    img_bg_blur = cv2.bitwise_and(img_orig_blur, img_bg_mask)
    img_concat = np.where(img_mask_color==255, img_orig, img_bg_blur)
    axarr[1, 1].imshow(cv2.cvtColor(img_concat, cv2.COLOR_BGR2RGB))
    axarr[1, 1].set_title('Blurred Image')
    
    # Return
    plt.tight_layout()
    plt.show()
```

```python
bgblur()
```

```python
import warnings
warnings.filterwarnings('ignore')

for i in range(3, 8):
    bgblur(os.getenv('HOME') + f'/aiffel/human_segmentation/images/{i}.jpg')
```

## 보완점

3번 사진의 뒤편에 있는 사람들까지 인식하는 문제에 대해서, 피사체를 특정하고 그것을 바탕으로 segmentation을 진행하는 아키텍쳐가 있으면 좋을 듯 하다.
예를 들어, 
1. 카메라가 찍고자 하는 피사체를 특정하고 바운더리를 지정하는 네트워크를 거치고
2. 찍지 않고자 하는 물체를 특정해 패널티 바운더리를 지정,
3. 피사체 바운더리 내에서 segmentation 진행,
3. segmentation과 패널티 바운더리가 겹치면 모델에 패널티 부여

```python
# Clear variables
%reset -f

# Import required libraries
import os
import urllib
import cv2
import numpy as np
from pixellib.semantic import semantic_segmentation
from matplotlib import pyplot as plt
```

```python
img_path=os.getenv('HOME')+'/aiffel/human_segmentation/images/face.jpg'
bg_img_path = os.getenv('HOME') + '/aiffel/human_segmentation/images/classroom.jpg'

# Create grid for plots
f, axarr = plt.subplots(2,2)

# Set and print reference image 
img_orig = cv2.imread(img_path)
axarr[0, 0].imshow(cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB))
axarr[0, 0].set_title('Reference Image')

# Get pre-trained model
model_dir = os.getenv('HOME') + '/aiffel/human_segmentation/models' 
model_file = os.path.join(model_dir, 'deeplabv3_xception_tf_dim_ordering_tf_kernels.h5') 
model_url = 'https://github.com/ayoolaolafenwa/PixelLib/releases/download/1.1/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5' 
if not os.path.exists(model_file):
    urllib.request.urlretrieve(model_url, model_file)
model = semantic_segmentation()
model.load_pascalvoc_model(model_file)
segvalues, output = model.segmentAsPascalvoc(img_path)

# Declare label names and predict on image
LABEL_NAMES = [
    'background', 'aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus',
    'car', 'cat', 'chair', 'cow', 'diningtable', 'dog', 'horse', 'motorbike',
    'person', 'pottedplant', 'sheep', 'sofa', 'train', 'tv'
]
axarr[0, 1].imshow(output)
axarr[0, 1].set_title('Predicted Segments')
# 디버깅 과정 기록
# Predict 값을 출력하기 위해 output이라는 변수이름을 사용했는데 나중에
# np.all(output==seg_color, axis=-1)에서 output을 변수이름으로 읽는 일이 일어나 시간이 많이 소요되었다.
out = ', '.join(f'"{LABEL_NAMES[class_id]}"' for class_id in segvalues['class_ids'])
print(f'This image contains {out}.')

# Create mask
seg_color = (128,128,192)
seg_map = np.all(output==seg_color, axis=-1) 
img_show = img_orig.copy()
img_mask = seg_map.astype(np.uint8) * 255
color_mask = cv2.applyColorMap(img_mask, cv2.COLORMAP_JET)
img_show = cv2.addWeighted(img_show, 0.6, color_mask, 0.4, 0.0)
axarr[1, 0].imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
axarr[1, 0].set_title('Masked Image')

# Replace background
bg_img = cv2.imread(bg_img_path)
bg_img = cv2.resize(bg_img, (img_orig.shape[1], img_orig.shape[0]))

# Create background mask
img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)
img_bg_mask = cv2.bitwise_not(img_mask_color)

# Combine foreground with new background
img_bg_replacement = cv2.bitwise_and(bg_img, bg_img, mask=img_bg_mask[:,:,0])
img_fg = cv2.bitwise_and(img_orig, img_orig, mask=img_mask[:,:,None].astype(np.uint8))
img_concat = cv2.add(img_bg_replacement, img_fg)

axarr[1, 1].imshow(cv2.cvtColor(img_concat, cv2.COLOR_BGR2RGB))
axarr[1, 1].set_title('Background Replaced Image')

# Return
plt.tight_layout()
plt.show()
```
